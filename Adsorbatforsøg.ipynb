```python
import numpy as np
from ase.build import molecule

# Get 'CH3O' from the G2 set
atoma = molecule('CH3O')

# Access atomic symbols and positions
symbols = atoma.symbols
coordinates = atoma.positions

# Find the index of the oxygen atom
oxygen_index = symbols.index('O')

# Get the oxygen coordinates
oxygen_coordinates = coordinates[oxygen_index]

# Subtract oxygen coordinates from all other atoms
adjusted_coordinates = coordinates - oxygen_coordinates

# Define the desired surface angle (in degrees)
desired_surface_angle = 63.0

# Calculate the rotation matrix for the desired angle around the y-axis
rotation_angle = np.radians(90.0 - desired_surface_angle)
rotation_matrix = np.array([[np.cos(rotation_angle), 0, np.sin(rotation_angle)],
                            [0, 1, 0],
                            [-np.sin(rotation_angle), 0, np.cos(rotation_angle)]])

# Rotate the entire adsorbate, including oxygen and carbon
rotated_coordinates = np.dot(adjusted_coordinates, rotation_matrix.T)

# Subtract oxygen coordinates from the rotated coordinates to set oxygen as origo
final_coordinates = rotated_coordinates - rotated_coordinates[oxygen_index]

# Print atomic symbols and final coordinates
for symbol, coord in zip(symbols, final_coordinates):
    print(f"{symbol}: {coord}")
```

```python
from ase.collections import g2
from ase.build import molecule
from ase.geometry.rotations import rotate_point
import numpy as np

# Get 'CH3O' from the G2 set
atoma = molecule('CH3O')
atoma.rotate
# Access atomic symbols and positions
symbols = atoma.symbols
coordinates = atoma.positions

# Find the index of the oxygen atom
oxygen_index = symbols.index('O')

# Get the oxygen coordinates
oxygen_coordinates = coordinates[oxygen_index]

# Subtract oxygen coordinates from all other atoms
adjusted_coordinates = coordinates - oxygen_coordinates

# Define the desired surface angle (in degrees)
desired_surface_angle = 63.0

# Calculate the rotation axis (normal to the surface)
rotation_axis = np.cross([1, 0, 0], oxygen_coordinates)

# Calculate the rotation angle needed to align with the surface angle
rotation_angle = np.radians(90.0 - desired_surface_angle)

# Rotate the entire adsorbate, including oxygen and carbon
rotated_coordinates = rotate_point(adjusted_coordinates, rotation_axis, rotation_angle)

# Subtract oxygen coordinates from the rotated coordinates to set oxygen as origo
final_coordinates = rotated_coordinates - rotated_coordinates[oxygen_index]

# Print atomic symbols and final coordinates
for symbol, coord in zip(symbols, final_coordinates):
    print(f"{symbol}: {coord}")
```
